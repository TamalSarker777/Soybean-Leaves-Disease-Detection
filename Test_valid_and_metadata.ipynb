```python
import numpy as np
import pandas as pd
import os
import shutil
import random
import glob
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

```python
import os
initial_count=0
total_count = 0
dir = "E:/Soybean/Soybean Original Dataset/Caterpillar"
for path in os.listdir(dir):
    if os.path.isfile(os.path.join(dir, path)):
        initial_count += 1
print(initial_count)
total_count+=initial_count
valid_caterpillar=int ((initial_count*20)/100)
print(valid_caterpillar)

initial_count = 0
dir = "E:/Soybean/Soybean Original Dataset/Diabrotica speciosa"
for path in os.listdir(dir):
    if os.path.isfile(os.path.join(dir, path)):
        initial_count += 1
print(initial_count)
total_count+=initial_count
valid_Diabrotica_speciosa=int ((initial_count*20)/100)
print(valid_Diabrotica_speciosa)


initial_count = 0
dir = "E:/Soybean/Soybean Original Dataset/Healthy"
for path in os.listdir(dir):
    if os.path.isfile(os.path.join(dir, path)):
        initial_count += 1
print(initial_count)
total_count+=initial_count
valid_Healthy=int ((initial_count*20)/100)
print(valid_Healthy)

print(total_count)
```

```python
# Organize data into train, valid dirs
os.chdir('E:/Soybean/Soybean Original Dataset')
if os.path.isdir('train') is False: 
    os.mkdir('train')
    os.mkdir('valid')
    list_folder=["Caterpillar","Diabrotica speciosa","Healthy"]
    for i in list_folder:
        shutil.move(f'{i}', 'train')
        os.mkdir(f'valid/{i}')

        if(i=="Healthy"):
          valid_samples = random.sample(os.listdir(f'train/{i}'),valid_Healthy)
          for j in valid_samples:
            shutil.move(f'train/{i}/{j}', f'valid/{i}')
        elif(i=="Caterpillar"):
          valid_samples = random.sample(os.listdir(f'train/{i}'), valid_caterpillar)
          for j in valid_samples:
            shutil.move(f'train/{i}/{j}', f'valid/{i}')
        else:
          valid_samples = random.sample(os.listdir(f'train/{i}'), valid_Diabrotica_speciosa)
          for j in valid_samples:
            shutil.move(f'train/{i}/{j}', f'valid/{i}')

os.chdir('../..')
```

```python
# Get the list of all files and directories
path = "E:/Soybean/Soybean Original Dataset/train/Caterpillar"
dir_list = os.listdir(path)
df = pd.DataFrame()
value= {
    'file name':dir_list,
    'healthy' :0,
    'Diabrotica speciosa' :0,
    'caterpillar':1
          }
df = pd.DataFrame(value)
df
```

```python
path = "E:/Soybean/Soybean Original Dataset/train/Diabrotica speciosa"
dir_list = os.listdir(path)
 
df2 = pd.DataFrame()
value= {
    'file name':dir_list,
    'healthy' :0,
    'Diabrotica speciosa' :1,
    'caterpillar':0
          }

df2 = pd.DataFrame(value)
df2
```

```python
path = "E:/Soybean/Soybean Original Dataset/train/Healthy"
dir_list = os.listdir(path)
 
df3 = pd.DataFrame()
value= {
    'file name':dir_list,
    'healthy' :1,
    'Diabrotica speciosa' :0,
    'caterpillar':0
          }

df3 = pd.DataFrame(value)
df3
```

```python
frames = [df, df2,df3]
  
train_frames = pd.concat(frames)
display(train_frames)

train_frames.to_csv('train.csv', index=False)
```
